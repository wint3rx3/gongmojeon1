```python
import pandas as pd
import asyncio
import aiohttp
from aiolimiter import AsyncLimiter
from dotenv import load_dotenv
import os
from tqdm.asyncio import tqdm_asyncio
import nest_asyncio

# ì¤‘ì²© ë£¨í”„ í—ˆìš©
nest_asyncio.apply()

# .envì—ì„œ API í‚¤ ë¶ˆëŸ¬ì˜¤ê¸°
load_dotenv()
KAKAO_API_KEY = os.getenv("KAKAO_API_KEY")
HEADERS = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}

limiter = AsyncLimiter(max_rate=5, time_period=1.0)

async def fetch_coords(session, address):
    async with limiter:
        try:
            url = "https://dapi.kakao.com/v2/local/search/address.json"
            params = {"query": address}
            async with session.get(url, headers=HEADERS, params=params) as resp:
                data = await resp.json()
                if data.get("documents"):
                    x = data["documents"][0]["x"]
                    y = data["documents"][0]["y"]
                    return address, x, y
                else:
                    return address, None, None
        except Exception as e:
            print(f"ì˜¤ë¥˜: {address} â†’ {e}")
            return address, None, None

async def main():
    df = pd.read_csv("data\ì„¸ì¢…íŠ¹ë³„ìì¹˜ì‹œ_ê³µê³µë°ì´í„°(2).csv", encoding="utf-8")
    addresses = df['adress'].fillna("").tolist()
    results = []

    async with aiohttp.ClientSession() as session:
        tasks = [fetch_coords(session, addr) for addr in addresses if addr.strip() != ""]
        for result in tqdm_asyncio.as_completed(tasks, total=len(tasks)):
            results.append(await result)

    coord_df = pd.DataFrame(results, columns=["adress", "x", "y"])
    df = df.merge(coord_df, on="adress", how="left")
    df.to_csv("data\ì„¸ì¢…íŠ¹ë³„ìì¹˜ì‹œ_ê³µê³µë°ì´í„°(3).csv", index=False, encoding="utf-8")
    print("ì¢Œí‘œ ë³‘ë ¬ ì²˜ë¦¬ ì™„ë£Œ!")

# Notebookì´ë‚˜ Colabì—ì„œëŠ” asyncio.run ëŒ€ì‹  await ì‚¬ìš©
await main()
```

```python
import pandas as pd

# 1. ë‘ íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_csv("data/ì„¸ì¢…íŠ¹ë³„ìì¹˜ì‹œ_ê³µê³µë°ì´í„°(3).csv", encoding="utf-8")
merged_df = pd.read_csv("output/ë²„ìŠ¤ì¶”ê°€(1).csv", encoding="utf-8")

# 2. ê³µê³µë°ì´í„°ì— id ì»¬ëŸ¼ ì¶”ê°€ (ì˜ˆ: ìë™ ë¶€ì—¬)
df['id'] = ['sj_%04d' % i for i in range(len(df))]

# 3. ì»¬ëŸ¼ ìˆœì„œë¥¼ ë§ì¶”ê¸°
df = df[['id', 'name', 'category', 'x', 'y']]  # 'adress'ëŠ” ì œì™¸
merged_df = merged_df[['id', 'name', 'category', 'x', 'y']]  # í˜¹ì‹œ ëª¨ë¥´ë‹ˆ í™•ì¸

# 4. í–‰ ë°©í–¥ ë³‘í•©
combined_df = pd.concat([merged_df, df], ignore_index=True)
```

```python
# 1. x ì—´ì˜ NaN ê°œìˆ˜
nan_count = combined_df['x'].isna().sum()

# 2. ë¹ˆ ë¬¸ìì—´ ê°œìˆ˜
empty_count = (combined_df['x'].astype(str).str.strip() == '').sum()

# ì´ ê²°ì¸¡ì¹˜ ê°œìˆ˜
total_missing = nan_count + empty_count

print(f"NaN ê°œìˆ˜: {nan_count}")
print(f"ë¹ˆ ë¬¸ìì—´ ê°œìˆ˜: {empty_count}")
print(f"ì´ ê²°ì¸¡ì¹˜ ê°œìˆ˜: {total_missing}")

# 3. ê²°ì¸¡ì¹˜ ì œê±°
combined_df_cleaned = combined_df.dropna(subset=['x'])
combined_df_cleaned = combined_df_cleaned[combined_df_cleaned['x'].astype(str).str.strip() != '']
print(f"\nì œê±° í›„ ë°ì´í„° ìˆ˜: {len(combined_df_cleaned)}")
```

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import BallTree
from sklearn.metrics.pairwise import haversine_distances
import networkx as nx
import math

# 1. ë°ì´í„° ë¡œë“œ ë° ì¤€ë¹„
df = combined_df_cleaned.reset_index(drop=True)
coords = np.radians(df[['y', 'x']].astype(float).values)

# 2. ê³µí†µ ë¬¸ì ìˆ˜ í™•ì¸ í•¨ìˆ˜
def common_chars(s1, s2):
    return len(set(str(s1)) & set(str(s2)))

# 3. BallTree
tree = BallTree(coords, metric='haversine')
radius_km = 0.05
radius_rad = radius_km / 6371

# 4. ìœ ì‚¬ ìŒ íƒìƒ‰
similar_pairs = []

for i, row in df.iterrows():
    name1 = row['name']
    coord1 = coords[i].reshape(1, -1)
    indices = tree.query_radius(coord1, r=radius_rad)[0]

    for j in indices:
        if i >= j:
            continue

        name2 = df.loc[j, 'name']
        if common_chars(name1, name2) >= 3:
            dist_km = haversine_distances([coord1[0], coords[j]])[0, 1] * 6371
            similar_pairs.append((i, j, name1, name2, dist_km))

# 5. ìœ ì‚¬ POI ìŒ â†’ ê·¸ë˜í”„ë¡œ ì—°ê²°
G = nx.Graph()
G.add_edges_from([(i, j) for i, j, _, _, _ in similar_pairs])

components = list(nx.connected_components(G))

# 6. êµ°ì§‘ë³„ POI ì •ë³´ ë³´ê¸°
cluster_info = []

for idx, group in enumerate(components):
    group = list(group)
    for i in group:
        cluster_info.append({
            'cluster_id': idx,
            'index': i,
            'name': df.loc[i, 'name'],
            'category': df.loc[i, 'category'],
            'x': df.loc[i, 'x'],
            'y': df.loc[i, 'y'],
        })

# 7. êµ°ì§‘ ì •ë³´ë¥¼ DataFrameìœ¼ë¡œ ì €ì¥/ì¶œë ¥
cluster_df = pd.DataFrame(cluster_info).sort_values(by='cluster_id')

# ì €ì¥ (ì„ íƒ)
cluster_df.to_csv("output/ì¤‘ë³µêµ°ì§‘_ëª©ë¡.csv", index=False, encoding='utf-8')
print(f"ğŸ” ìœ ì‚¬ POIê°€ í¬í•¨ëœ êµ°ì§‘ ìˆ˜: {len(components)}ê°œ")
print("ğŸ“ ì¤‘ë³µ êµ°ì§‘ ì •ë³´ê°€ 'output/ì¤‘ë³µêµ°ì§‘_ëª©ë¡.csv'ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import BallTree

df_cleaned = pd.read_csv("output/ê³µê³µë°ì´í„°_POI_ì¹´í…Œê³ ë¦¬ì™„ì„±.csv", encoding="utf-8")

# ëˆ„ë½ëœ ë§¤í•‘ ë³´ì™„
SECOND_TO_FIRST['ì•½êµ­'] = 'ì•½êµ­'
FIRST_TO_CODE['ì•½êµ­'] = 'PM9'  # ì¹´ì¹´ì˜¤ë§µ ê¸°ì¤€ ì•½êµ­ ì½”ë“œ

# ë‹¤ì‹œ ëŒ€ë¶„ë¥˜ ë° ì½”ë“œ ì»¬ëŸ¼ ì—…ë°ì´íŠ¸
df_cleaned['category_1st'] = df_cleaned['category_2nd'].map(SECOND_TO_FIRST)
df_cleaned['category_code'] = df_cleaned['category_1st'].map(FIRST_TO_CODE)

# 1. ë°ì´í„° ì¤€ë¹„
df = df_cleaned.reset_index(drop=True)

# ì¢Œí‘œ ê²°ì¸¡ì¹˜ ì œê±° (í˜¹ì‹œ ë‚¨ì•„ìˆì„ ê²½ìš° ëŒ€ë¹„)
df = df[df['x'].notnull() & df['y'].notnull()].copy()
coords = np.radians(df[['y', 'x']].astype(float).values)

# 2. ì´ë¦„ ë“±ì¥ íšŸìˆ˜ ê³„ì‚°
name_counts = df['name'].value_counts()
multi_names = name_counts[name_counts > 1].index  # 2ë²ˆ ì´ìƒ ë“±ì¥í•œ ì´ë¦„ë§Œ

# 3. BallTree ì„¤ì •
tree = BallTree(coords, metric='haversine')
radius_km = 0.05  # 50m
radius_rad = radius_km / 6371  # ì§€êµ¬ ë°˜ì§€ë¦„ìœ¼ë¡œ í™˜ì‚°

# 4. ì¤‘ë³µ í›„ë³´ íƒìƒ‰
duplicate_indices = set()

for i, row in df.iterrows():
    name_i = row['name']
    if name_i not in multi_names:
        continue  # ë‹¨ 1íšŒ ë“±ì¥í•œ ì´ë¦„ì€ ë¬´ì‹œ

    coord_i = coords[i].reshape(1, -1)
    indices = tree.query_radius(coord_i, r=radius_rad)[0]

    for j in indices:
        if i >= j:
            continue
        if df.loc[j, 'name'] == name_i:
            duplicate_indices.add(j)  # ë‚˜ì¤‘ í•­ëª©ë§Œ ì œê±°

# 5. ì œê±° ìˆ˜í–‰
df_deduplicated = df.drop(index=duplicate_indices).reset_index(drop=True)
df_removed = df.loc[list(duplicate_indices)].reset_index(drop=True)

# 6. ì €ì¥
df_deduplicated.to_csv("output/ê³µê³µë°ì´í„°_POI_ì¤‘ë³µì œê±°_ì•ˆì „ì™„ë£Œ.csv", index=False, encoding="utf-8")
df_removed.to_csv("output/ì¤‘ë³µì œê±°ëœ_POI_ì•ˆì „ëª©ë¡.csv", index=False, encoding="utf-8")

# 7. ë¡œê·¸ ì¶œë ¥
print(f"ğŸ“¦ ì „ì²´ ì›ë³¸ ê°œìˆ˜: {len(df)}")
print(f"ğŸ—‘ï¸ ì œê±°ëœ ì¤‘ë³µ POI ìˆ˜: {len(df_removed)}")
print(f"âœ… ìµœì¢… ë‚¨ì€ POI ìˆ˜: {len(df_deduplicated)}")
```
